### 2b. Visualization with widgets to select year, depth, parameter, and color

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Download BATS Bottle Data from BCO-DMO using CSV link
url = "https://datadocs.bco-dmo.org/file/OJRrAA6hpQREv3/3782_v6_bats_bottle.csv"
df = pd.read_csv(url)
#df.head(10) # check first 10 rows of dataframe

# Create a function to update the plot based on the selected year range and parameter
def update_plot(year_range, depth_range, y_param, color):
    # Sequential filters on the dataset, year then depth set by two sliders
    filtered_df = df[(df['decimal_year'] >= year_range[0]) & (df['decimal_year'] <= year_range[1])]
    filtered_df2 = filtered_df[(filtered_df['Depth'] >= depth_range[0]) & (filtered_df['Depth'] <= depth_range[1])]
    plt.figure(figsize=(10, 6))
    plt.plot(filtered_df2['decimal_year'], filtered_df2[y_param], marker='.', linestyle='', color=color)
    plt.xlabel('Decimal Year')
    plt.ylabel(y_param)
    plt.title(f'{y_param} vs Decimal Year')
    # plt.grid(True) # adds gridlines
    plt.show()

# Create a slider to select the year range
year_slider = widgets.IntRangeSlider(
    value=[1988, 2024],
    min=int(df['decimal_year'].min()),
    max=int(df['decimal_year'].max()),
    step=1,
    description='Year Range:',
    continuous_update=False
)
depth_slider = widgets.IntRangeSlider(
    value=[0, 5000],
    min=int(df['Depth'].min()),
    max=int(df['Depth'].max()),
    step=1,
    description='Depth Range:',
    continuous_update=False
)
# Create a dropdown to select the parameter to plot on the Y axis
# Filter out columns that start with 'QF'
filtered_columns = [col for col in df.columns[1:] if not col.startswith('QF')]

param_dropdown = widgets.Dropdown(
    options=filtered_columns,  # Use the filtered columns
    description='Parameter:',
    disabled=False,
    value='Temp'  # Default to 'Temp' for the Y axis
)
# Create a dropdown to select the color of the plot
color_dropdown = widgets.Dropdown(
    options=['blue', 'green', 'purple', 'orange', 'pink', 'brown', 'cyan', 'magenta'],
    description='Color:',
    disabled=False,
    value='blue'  # Default to 'blue'
)
# Display the slider, dropdown and the plot
widgets.interact(update_plot, year_range=year_slider, depth_range=depth_slider, y_param=param_dropdown, color=color_dropdown)
```
